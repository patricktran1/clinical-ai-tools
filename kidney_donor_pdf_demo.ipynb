# Kidney Donor PDF Parsing Demo
This notebook demonstrates how to extract and clean text from a kidney donor intake form using PyMuPDF.

```
# Install PyMuPDF if not already installed
# !pip install pymupdf
import fitz  # PyMuPDF
import os
```

```
# Define text extraction function
def extract_text_from_pdf(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text
```

```
# Define basic cleaning function
def clean_extracted_text(raw_text):
    lines = raw_text.split('\n')
    lines = [line.strip() for line in lines if line.strip()]
    return "\n".join(lines)
```

```
# Run the parser on a sample PDF (replace path with actual file)
sample_pdf_path = 'sample_kidney_donor_form.pdf'  # Replace with your test file path

try:
    raw_text = extract_text_from_pdf(sample_pdf_path)
    cleaned_text = clean_extracted_text(raw_text)
    print("Cleaned Text Output:")
    print(cleaned_text)
except Exception as e:
    print(f"Error: {e}")
```

This cleaned output can now be passed into a large language model (LLM) or structured for further clinical analysis.
