# Cervical Cancer Risk Assessment Demo (LLM Prompt)
This notebook demonstrates how to use a structured prompt with an LLM to assess cervical cancer screening results.

```
# Example input (can be modified)
patient_info = {
    "age": 33,
    "pap_smear": "LSIL",
    "hpv_status": "Positive",
    "symptoms": "None"
}
```

```
# Construct prompt from structured data
def build_prompt(info):
    prompt = f"""You are a clinical decision support assistant for OB-GYNs.

Your goal is to help evaluate cervical screening results and recommend next steps. Based on the patient's profile, generate:

1. A risk level (Low / Moderate / High)
2. The appropriate follow-up action

Input:
- Patient Age: {info['age']}
- Pap Smear: {info['pap_smear']}
- HPV: {info['hpv_status']}
- Symptoms: {info['symptoms']}

Respond using current clinical guidelines (e.g., ASCCP) and provide a professional summary for physician reference.
"""
    return prompt

prompt = build_prompt(patient_info)
print(prompt)
```
